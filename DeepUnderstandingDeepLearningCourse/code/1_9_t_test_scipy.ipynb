```python
import numpy as np
import matplotlib.pyplot as plt

import scipy.stats as st
```

```python
n1 = 30 # num samples in dataset 1
n2 = 40 # num samples in dataset 2
x = 1 # mean of model 1
y = 2 # mean of model 2

# generate data
data1 = x + np.random.randn(n1)
data2 = y + np.random.randn(n2)

# plot
plt.plot(np.zeros(n1), data1, 'ro', markerfacecolor='w', markersize=14)
plt.plot(np.ones(n2), data2, 'bs', markerfacecolor='w', markersize=14)
plt.xlim([-1, 2])
plt.xticks([0,1], labels=['Group1', 'Group2'])
plt.show()
```

```python
# There are several varieties of t_test (different formulas)
t, p = st.ttest_ind(data1, data2) # ind = independent samples
print(t)
print(p)

# group 2 has larger mean than group 1
```

```python
# common way to show t-test results in a plot
fig = plt.figure(figsize=(10, 4))
plt.rcParams.update({'font.size': 12})

plt.plot(0+np.random.randn(n1)/30, data1, 'ro', markerfacecolor='w', markersize=14)
plt.plot(1+np.random.randn(n2)/30, data2, 'bs', markerfacecolor='w', markersize=14)
```
