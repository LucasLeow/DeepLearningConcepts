```python
import numpy as np
import matplotlib.pyplot as plt

import sympy as sym
import sympy.plotting.plot as symplot
```

```python
x = sym.symbols('x') # specify x as symbolic

fx = 2*x**2
df = sym.diff(fx, x)
print(fx)
print(df)
```

```python
symplot(fx, (x, -4, 4), title='fx')
symplot(df, (x, -4, 4), title='df')
plt.show()
```
